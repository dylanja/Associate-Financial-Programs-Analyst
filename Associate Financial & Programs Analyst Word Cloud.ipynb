```python
stopwords = set(STOPWORDS)
```

```python
pge1_wc = WordCloud(
    background_color='skyblue',
    max_words=2000,
    stopwords=stopwords)

pge1_wc.generate(pge1)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(pge1_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
